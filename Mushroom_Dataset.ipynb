### 4) Categorical 속성을 Numerical 속성으로 변환

```python
df[0] = df[0].map({'p' : 1, 'e' : 0})
df
```

- 모든 문자형 데이터들을 수치형 데이터로 변환

```python
num_columns = df.shape[1]
map_dic = {}
for i in range(num_columns):
    unique_array = df[i].unique()
    N = len(unique_array)
    map_dic[i] = {}
    for j in range(N):
        map_dic[i][unique_array[j]] = j
    df[i] = df[i].map(map_dic[i])    
df
```

- 모든 수치화 한 데이터를 다시 정규화 하는 과정

```python
for i in range(1, num_columns):
    unique_array = df[i].unique()
    N = len(unique_array)
    map_dic_sub = {}
    for j in range(N):
        if j == 0:
            map_dic_sub[j] = 0
        else:
            map_dic_sub[j] = j / float(N - 1)   #정규화를 위한 과정
    df[i] = df[i].map(map_dic_sub)
df
```

## Entropy

- Entropy 함수 (entropy의 값을 계산하는 함수)

```python
def entropy(labels):
    """ Computes entropy of label distribution. """
    from collections import Counter
    from math import log
    n_labels = len(labels)

    if n_labels <= 1:
        return 0

    counts = np.array(Counter(labels).values())
    probs = counts / float(n_labels)
    n_classes = np.count_nonzero(probs)

    if n_classes <= 1:
        return 0

    ent = 0.

    # Compute standard entropy.
    for i in probs:
        ent -= i * log(i, n_classes)   #과제시에는 Binary가 아닌 4개의 구분을 해야한다.

    return ent
```

## Information Gain (IG)

```python
parent_size = float(len(df))
parent_size
```

```python
for i in range(1, 23):
        sub_sum_entropy = 0.0
        print i
        for j in range(df_per_attr[i]['ngorups']):
            sub_size = len(df_per_attr[i]['subgroups'][j])
            sub_entropy = entropy([item[0] for item in df_per_attr[i]['subgroups'][j].values])
            print "%3.2f * %5.4f = %5.4f" % (sub_size / parent_size, sub_entropy, (sub_size / parent_size) * sub_entropy)
            sub_sum_entropy = sub_sum_entropy + (sub_size / parent_size) * sub_entropy
        ig = entropy(df[0].values) - sub_sum_entropy
        df_per_attr[i]['information_gain'] = ig
        print "Information Gain: %5.4f - %5.4f = %5.4f" % (entropy(df[0].values), sub_sum_entropy, ig)
        print
```

```python
import operator
information_gain_dict = {}
for i in range(1, 23):
    information_gain_dict[i] = df_per_attr[i]['information_gain']

information_gain_dict
```
