### [Mission 5] 단일변수 선형 회귀분석을 통하여 가장 영향이 높은 속성 선발
- 이전 문제에서 area의 값에 자연로그를 취하여 변환하는 것은 수행하지 마시오.
    - 원래 area 데이터에 대한 선형회귀 분석 수행
- 다음 요인들 중 area와 가장 연관성이 높은 주요 요인을 선정
    - FFMC
    - DMC
    - DC
    - ISI
    - temp
    - RH
    - wind
    - rain

```python
sdf = df[['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']]
sdf
```

```python
col = ['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain']
predicator_analysis = {}
for i in range(0, 8):
    predicator_analysis[i] = Series(np.empty(6), index=['slope', 'intercept', 'r_value', 'r_squared', 'p_value', 'std_err'])
    predicator_analysis[i][0],\
    predicator_analysis[i][1],\
    predicator_analysis[i][2],\
    predicator_analysis[i][4],\
    predicator_analysis[i][5] = stats.linregress(sdf[col[i]].values, sdf['area'].values)
    predicator_analysis[i][3] = predicator_analysis[i][2] ** 2

df2 = pd.DataFrame(predicator_analysis)
df2.columns = col
df2
```

- p-value의 값이 0.05보다 낮은 값이 가장 연관성이 높은 요인으로 보았다.
    - 따라서 **temp**값이 가장 연관성이 높은 주요 요인으로 선정

### [Mission 6] 가장 영향이 높은 요인 두 개로 다변수 선형 회귀분석 수행
- 전체 517개의 데이터에 대한 area 값과 선형회귀 분석 결과인 predicted area 값을 모두 한꺼번에 출력하여 비교하시오.
- 선형 회귀분석 성능이 좋은지 자신의 의견을 제시하시오.

- 지난번 수업시간에 들은 다변수 선형 회귀분석을 이용하여 분석을 함.
- sklearn 모듈을 활용한 다중 변수 선형 회귀

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

- 위에 p-value 값들 중에서 0.05보다 작거나 혹은 0.05에 가까운 두 가지 값을 선택
    - temp (0.05보다 작다, 0.026101)
    - RH (0.05보다 크며, 0.086271)

```python
X = zip(sdf['temp'], sdf['RH'])
y = sdf['area'].values
```

```python
regr = regr.fit(X, y)   #temp와 RH값을 area와 비교하여 각각의 계수를 구함

print 'Coefficients:', regr.coef_
print 'Intercept:', regr.intercept_
```

```python
# Plot outputs
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(17, 7))
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(sdf['temp'], sdf['RH'], sdf['area'])

ax2 = fig.add_subplot(122, projection='3d')
ax2.scatter(sdf['temp'], sdf['RH'], sdf['area'])
# create x,y
xx, yy = np.meshgrid(range(int(sdf['temp'].min()), int(sdf['temp'].max())), range(int(sdf['RH'].min()), int(sdf['RH'].max())))
# calculate corresponding z
z = 1.9288 + 0.8810 * xx - 0.1292 * yy
ax2.plot_surface(xx, yy, z, rstride=1, cstride=1, linewidth=0, color="yellow", shade=False)
```

- 다음과 같이 결과가 나왔으며, temp의 값의 경우 p-value의 값이 작기 때문에 평가하기 좋은 지표가 되는듯 하다.
- 하지만 RH 값의 경우 0.05보다 큰 값이기 때문에 비교의 값으로 좋은것 같지는 않다.

### [Mission 7] 가장 영향이 높은 요인 두 개로 다변수 로지스틱 회귀분석 수행

- 두 가지 부류로 나누는 기준은 log(1 + area) 값이 50% percentile 이상인 것과 50% percentile 이하인 것으로 정함
    - log(1 + area) 값이 50% percentile 이상이면 'fire' 컬럼 값이 1
    - log(1 + area) 값이 50% percentile 이하이면 'fire' 컬럼 값이 0

```python
sdf = df[['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']]
sdf[['area']] = np.log1p(sdf[['area']])
sdf
```

```python
sdf['fire'] = sdf['area'].apply(lambda x: 1 if x >= 0.418710 else 0)
sdf
```

- 선형 회귀분석에서 나왔듯이 p-value의 값이 가장 작았던 값을 가지고 다변수 로지스틱 회귀분석 수행
    - temp
    - RH

```python
sdf_data = sdf[['temp', 'RH']]   #두개의 데이터만 추려냄
sdf_target = sdf[['fire']]       #fire의 값을 타겟으로 추려냄
```

```python
sdf_data
```

```python
predicted = []
for i in range(0, len(sdf_data)):
    predicted.append(regr.predict(sdf_data.values[[i]]))

sdf.ix[:, 'predicted'] = predicted
sdf
```

```python
correctly_predicted = 0
for i in range(0, len(sdf_data)):
    if predicted[i] == sdf_target.values[[i]]:
        correctly_predicted += 1
        
print float(correctly_predicted) / len(sdf_data)
```

- 다변수 로지스틱 회귀분석을 수행한 결과 정확도가 낮은것 같았다.
- 이유는 RH의 값의 영향으로 그런것 같다.
