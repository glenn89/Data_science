## 1. Graph Generators를 사용한 그래프 특성 분석

```python
import networkx as nx
import matplotlib.pyplot as plt
import networkx.algorithms as algo
%matplotlib inline
```

- 각 그래프의 radius, diameter, center, periphery, clustering의 값을 나타내는 함수

```python
def result(g):   #각 그래프의 값들을 보여주는 함수 정의
    print "radius(g):", algo.radius(g)
    print "diameter(g):", algo.diameter(g)
    print "center(g):", algo.center(g)
    print "periphery(g):", algo.periphery(g)
    print "clustering(g):",nx.clustering(g)
    print "Average Clustering Coefficient(g):",sum(nx.clustering(g).values())/len(nx.clustering(g))
```

### Classic으로 분류된 그래프 중 5개 선택

- Barbell 그래프

```python
g= nx.barbell_graph(5,2)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

- Complete 그래프

```python
g = nx.complete_graph(8)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

- Complete_bipartite 그래프

```python
g =nx.complete_bipartite_graph(2,8)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

- Circular_ladder 그래프

```python
g = nx.circular_ladder_graph(5)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

- Cycle 그래프

```python
g = nx.cycle_graph(10)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

#### Classic 그래프 고찰
- **barbel** 그래프의 경우 바벨의 모양과 같기 때문에 양 끝은 클러스터를 형성 하고 있다. 따라서 2개의 클러스터가 생성이 되게 된다. 또한 센터는 중앙에 두 클러스터를 이어주는 5번 6번 노드가 센터가 된다. 클러스터의 평균 값은 두 클러스터의 값이 평균이 된다.
- **Complete** 그래프의 경우 서로가 모두가 완전한 그래프가 된다. 그렇기 때문에 클러스터의 평균, 값은 다 1이 된다. 또한 서로가 다 연결 되어 있기 때문에 diameter 값도 1이 되게 된다.
- **Complete_bipartite** 그래프의 경우 이분 그래프이기 때문에 서로가 완전그래프를 이루는 경우가 없다. 따라서 클러스터의 값은 0이 된다. 하지만 각 노드들이 서로 하나의 노드만 거친다면 서로가 통신?을 할 수 있기 때문에 center와 periphery 값이 같게 된다.
- **circular_ladder** 그래프의 경우 이쁘게 하나의 입체 도형을 만들었다. 이 그래프 또한 cluster를 형성하지는 않기 때문에 위의 완전 이분 그래프와 같은 값이 나온다. 하지만 입체 도형이라서 그런지 radius와 diameter의 값은 3이 된다.
- **cycle** 그래프는 각 노드들이 하나의 원형을 이루게 된다. 각 노드들이 하나의 선 상에 위치하게 되기 때문에 위에 나온 값들을 같게 된다.

### Random으로 분류된 그래프 중 5개 선택

- Fast_gnp 랜덤 그래프

```python
g = nx.fast_gnp_random_graph(10, 0.7)   #차례로 노드의 갯수, 노드간에 edge의 생성 확률
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

- dense_gnm 랜덤 그래프

```python
g = nx.dense_gnm_random_graph(10, 20)   #차례로 노드의 수, 엣지의 갯수
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

- erdos_renyi 랜덤 그래프

```python
g = nx.erdos_renyi_graph(10, 0.7)   #차례로 노드의 수, 엣지 생성 확률
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

- barabasi_albert 랜덤 그래프

```python
g = nx.barabasi_albert_graph(10, 2)
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

- random_regular 그래프

```python
g = nx.random_regular_graph(4, 6)   #차례로 degree 갯수, 노드의 갯수
nx.draw(g, node_size=200, with_labels=True, font_size=10)

result(g)
```

#### Random 그래프 고찰
- **fast_gnp** 그래프는 한 노드가 다른 노드에 엣지를 생성하는 확률을 주어서 그래프를 생성하는 것이다. 따라서 확률을 다르게 주면 줄수록 그래프의 각 값들이 변하게 된다.
- **dense_gnm** 노드와 엣지의 갯수가 주어지면 랜덤하게 노드와 엣지의 갯수를 보고 그래프를 생성하게 되는 것이다. 하지만 fast_gnp 그래프와는 다르게 클러스터가 0이 되는 값은 안나오는것 같다.
- **erdos_renyi** 그래프는 위의 fast_gnp 그래프와 큰 차이를 잘 알지 못하겠다.
- **barabasi_albert** 그래프는 척도없는 네트워크라고 불리우게 된다. 그러므로 한노드가 평균 이상의 엣지를 가지게 되는 경우를 나타내는 그래프이다. 보통은 네트워크의 모양을 나타내는 그리프인것 같다. 따라서 새로운 노드가 생기면 두 개의 링크를 가지되, 무작위가 아니라 선호에 따라 연결하는 노드이다.
- **random_regular** 그래프에서 노드에 붙는 엣지의 수를 결정하면 랜덤하게 엣지의 갯수에 따라 노드가 연결이 되는 일반적인 랜덤 그래프의 모습이다.
