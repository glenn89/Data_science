# Decision Tree

- iris 데이터를 이용하여 Decistion tree를 생성
- scikit이라고 하는 Machine-learning 모듈을 이용하여 학습 한다.

```python
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.externals.six import StringIO

iris = load_iris()   #iris데이터는 기본적으로  python 에서 함수로 제공이 되고 있다. (유명하기 때문)
print type(iris)
```

```python
iris.data
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')   #아직 학습이 되지 않은 데이터
clf = clf.fit(iris.data, iris.target)   #fit을 하는 순간 디시전 트리 모델이 생성
```

```python
with open("iris.dot", 'w') as f:
    tree.export_graphviz(clf, out_file=f)   #dot이라는 언어를 사용하여 트리로 나타낼 수 있다.
```
